# MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
# to display matplotlib in jupyter notebook to avoid plot.show()
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
# plotting using matplotlib
## plt scatter
plt.scatter(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
```

```python
y = x*x
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D for y = x^2')
```

```python
plt.savefig('2DGraph.png')
```

```python
plt.plot(x,y,'r*--', linewidth = 1, markersize = 10)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Plot Example")
```

```python
# subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(a,b,'g')
plt.subplot(2,2,3)
plt.plot(x,b,'ko')
```

```python
x = np.arange(0,4* np.pi, 0.01)
y = np.sin(x)
plt.title("Sine Function")
plt.plot(x,y,'--')
```

```python
#Bar plot
x = [2,8,10]
y = [11,16,9]

plt.bar(x,y)
```

## Box plot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

## pie chart

```python
labels = 'Nokia', 'Samsung', 'Xiomi', 'Realme'
sizes = [215,130,245,210]
colors = ['gold','green','lightcoral','red']
explode = (0.4,0,0,0)

#plot
plt.pie(sizes,explode=explode,labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
