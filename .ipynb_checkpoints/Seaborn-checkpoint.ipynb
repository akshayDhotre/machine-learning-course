# Seaborn

## Distribution plots

```python
import seaborn as sns
```

```python
#sample dataset provided by seaborn
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap
A correlation heatmap uses colored cells, monochrome scale, to show 2d correlation matrix between two dimensions or event types. It is very importnant in feature selection.

```python
df.corr() ## compares only numeric values i.e. floating values or integer values
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

## Joint Plot
relation between two numeric values

### Bivariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

## Pair Polt

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Distplot - for histogram

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=40)
```

## Categorial Plots

```python
## count plot
sns.countplot('day',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

## BoxPlot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df, palette='rainbow')
```

```python
sns.boxplot(orient='v',data=df)
```

```python
sns.boxplot(y='sex',x='total_bill',data=df, hue='smoker')
```

## Violin plot

```python
sns.violinplot(y='day',x='total_bill',data=df, hue='smoker')
```
